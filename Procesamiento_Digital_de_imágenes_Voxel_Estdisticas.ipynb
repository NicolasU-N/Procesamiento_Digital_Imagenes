# Procesamiento Digital de Imágenes: Estadísticas, Histograma y vóxel

```
from google.colab import files
from IPython.display import Image
```

```
# Load image
import cv2
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
```

```
img_in = files.upload()
```

```
img=mpimg.imread('2.jpg') 
imgplot = plt.imshow(img)
print('Dimensions : ',img.shape) 
print()
```

## Imágen Blanca

```
# Calculamos la media de cada canal 
channels = cv2.mean(img)

# Intercambio valores azules y rojos (haciéndolo RGB, no BGR)
observation = np.array([(channels[2], channels[1], channels[0])])

# Mostrar valores medios del canal
print(observation)

#Ver colores medios de la imagen
plt.title("R                        G                       B")
plt.imshow(observation), plt.axis("off")
plt.show()

channels1 = cv2.meanStdDev(img)

print(channels1)
print()
# Desviacion estandar y media total

desviation = img.std()
print('Desviación Estándar: ')
print(desviation)
mean = img.mean()
print('Media: ')
print(mean)
```

```
color = ('b','g','r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim(0,256)


plt.xlabel('Channel value')
plt.ylabel('Frequency')
plt.text(160, 5330, r'$\mu=237.99,\ \sigma=31.42$')
plt.grid(True)
plt.show()
```

## Resize

```
 
scale_percent = 10.2 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
imgplot = plt.imshow(resized) 
plt.show() 


desviation = resized.std()
print('Desviación Estándar: ')
print(desviation)
mean = resized.mean()
print('Media: ')
print(mean)
```

```
color = ('b','g','r')

for i, col in enumerate(color):
    histr = cv2.calcHist([resized],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim(0,256)



plt.xlabel('Channel value')
plt.ylabel('Frequency')
plt.axis([0, 256, 0, 280])
plt.text(151, 110, r'$\mu=237.99,\ \sigma=29.48$')
plt.grid(True)
plt.show()
```

**VOXEL**

```
image_data = np.asarray(resized)

for i in range(len(image_data)):
    for j in range(len(image_data[0])):             
        flag=image_data[i,j]
        plt.plot(flag)

plt.xlabel('N° Canal')
plt.ylabel('Intensidad')
plt.grid(True)
plt.show()
        
```

# **Procesamiento de imágenes con numpy y opencv**

Para esta práctica se utilizaron 2 imágenes de un objeto con color uniforme, capturadas con una camara digital.

La **primera** imagen se capturo en un ambiente interior con condiciones de luz controladas.

La **segunda** fue tomada en un espacio abierto con iluminación natural en el atardecer.

- Para cada imagen se separaron los respectivos canales *R G B* 
- Se graficó el histograma que me muestra la distribución de las intensidades de píxeles en la imagen.
- Por último se compararon las dos imágenes y se representaron dichas diferencias por canales en escalas de grises.

```
import numpy as np

import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
```

## **Primera Imagen**

```
img=mpimg.imread('image_1_m (1).jpeg') 
imgplot = plt.imshow(img) 
plt.show() 
print(img.shape)


#Canal Rojo

image_red=np.copy(img) # creo una copia de la imagen para preservar la original
image_red[:,:,1]=0
image_red[:,:,2]=0
plt.title("Canal rojo")
imgplot = plt.imshow(image_red)
plt.show()

#Canal Verde

image_green=np.copy(img) # creo una copia de la imagen para preservar la original
image_green[:,:,2]=0
image_green[:,:,0]=0
plt.title("Canal verde")
imgplot = plt.imshow(image_green)
plt.show()

#Canal Azul

image_blue=np.copy(img) # creo una copia de la imagen para preservar la original
image_blue[:,:,1]=0
image_blue[:,:,0]=0
plt.title("Canal azul")
imgplot = plt.imshow(image_blue)
plt.show()
```

```
#Histograma imagen en interior

# Importando libreria opencv
import cv2

color = ('b','g','r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim(0,256)

plt.show()
```
